# Part 3, 4: Combined Geometric and Appearance Model with Principal Component Analysis

1. Load images.
2. Load landmarks.
3. Compute mean landmarks.
4. Warp all images from their landmarks to mean landmarks.
5. Normalize landmarks (mean center, divide by standard deviation)
6. Find eigenvectors of normalized landmarks covariance matrix.
7. Project test set landmarks onto landmark eigenvectors and obtain compressed test set geometry cofficients.
8. Compute mean warped images.
9. Normalize warped images (mean center, divide by standard deviation)
10. Find eigenvectors of normalized warped images covariance matrix.
11. Project test set warped images onto warped image eigenvectors and obtain compressed test set appearance coefficients.
12. Reconstruct test set warped images using appearance coefficients and appearance eigenvectors.
13. Reconstruct test set original landmarks using geometry coefficients and geometry eigenvectors.
14. Warp reconstructed test set warped images from mean landmarks to reconstructed landmarks.
15. Evaluate reconstruction.

Import our wonderful dependencies that make this all happen.

## Warping

Now we'll need some functions. The next function generates a list of points arranged in a grid.

```python
def generate_grid_points(xs, ys):
    m = np.zeros((len(xs)*len(ys), 2))
    for i in range(len(ys)):
        m[len(xs)*i:len(xs)*(i+1), 0] = xs
        m[len(xs)*i:len(xs)*(i+1), 1] = np.ones((len(xs),)) * ys[i]
    return m
```

This function will warp an arbitrary number of landmark points which we'll use to geometrically canonize the images.

```python
def warp_image(image, image_shape, from_landmarks, to_landmarks, anchor_borders=True, border_segments=2, interpolation_method='linear'):
    
    if anchor_borders:
        segments = border_segments
        borders = np.zeros((segments * 4, 2))
        borders[segments*0:segments*1, 0] = np.linspace(0, 1, segments, False) * image_shape[0]
        borders[segments*0:segments*1, 1] = 0
        borders[segments*1:segments*2, 0] = image_shape[0]
        borders[segments*1:segments*2, 1] = np.linspace(0, 1, segments, False) * image_shape[1]
        borders[segments*2:segments*3, 0] = (1 - np.linspace(0, 1, segments, False) ) * image_shape[0]
        borders[segments*2:segments*3, 1] = image_shape[1]
        borders[segments*3:segments*4, 0] = 0
        borders[segments*3:segments*4, 1] = (1 - np.linspace(0, 1, segments, False) ) * image_shape[1]
        border_deltas = np.zeros((segments * 4, 2))
        
    grid_points = generate_grid_points(np.arange(image_shape[0]), np.arange(image_shape[1]))
        
    deltas = to_landmarks - from_landmarks
    if anchor_borders:
        interp_points = np.vstack((from_landmarks, borders))
        deltas = np.concatenate((deltas, border_deltas))
    else:
        interp_points = from_landmarks
        deltas = deltas
        
    interp_delta = sp.interpolate.griddata(interp_points,
                                           deltas,
                                           grid_points,
                                           method=interpolation_method,
                                           fill_value=0)
    
    orig_pixels = grid_points
    disp_pixels = orig_pixels + interp_delta
    warped_image = sp.interpolate.griddata(disp_pixels,
                                           image,
                                           orig_pixels,
                                           method=interpolation_method,
                                           fill_value=0)
    
    return warped_image
```

## Landmark Principal Component Analysis

Now we'll do principal component analysis on the geometry. First we'll center our landmarks on the mean and normalize them by the standard deviation.

```python
norm_landmarks = np.zeros(training_landmarks.shape)
std_landmark = 1#np.std(training_landmarks)
for i in range(num_of_training_samples):
    norm_landmarks[i, :] = (training_landmarks[i, :] - mean_landmark) / std_landmark
```

At this point we would create a covariance matrix to get our principal components.

```python
geometric_covariance = [None] * 2
for i in range(2):
    geometric_covariance[i] = norm_landmarks[:, :, i].T.dot(norm_landmarks[:, :, i])
```

```python
plt.figure(figsize(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(geometric_covariance[0])
plt.title('$x$ Covariance')
plt.colorbar()
plt.subplot(1, 2, 2)
plt.imshow(geometric_covariance[1])
plt.title('$y$ Covariance')
plt.colorbar()
```

Find the eigenvectors of the covariance matrix.

```python
geometric_eig_vals = np.zeros((2, num_of_landmarks))
geometric_eig_vecs = np.zeros((2, num_of_landmarks, num_of_landmarks))
for i in range(2):
    junk, geometric_eig_vals[i], geometric_eig_vecs[i] = np.linalg.svd(geometric_covariance[i])
```

Normalize the eigenvectors because I want to. Their length isn't important since we are just going to project onto them anyway. Also, if they're normalized then they form an orthonormal bases which can be treated as a rotation.

```python
for i in range(2):
    geometric_eig_vecs[i] = np.apply_along_axis(lambda x: x/np.linalg.norm(x), 1, geometric_eig_vecs[i])
```

We're compressing so we'll use a truncated codebook.

```python
num_of_geometric_bases = 10
trunc_geometric_eig_vecs = geometric_eig_vecs[:, :num_of_geometric_bases, :]
```

Plot the first $100$ eigenvalues.

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(geometric_eig_vals[0][:100])
plt.title('$x$ Eigenvalues')
plt.subplot(1, 2, 2)
plt.plot(geometric_eig_vals[0][:100])
plt.title('$y$ Eigenvalues')
plt.show()
```

## Test Reconstruction

First we'll need some error metrics.

```python
def ncc(a, b):
    return np.sum( (a - np.mean(a)).dot(b - np.mean(b)) / (np.std(a) * np.std(b)) ) / len(a)
```

```python
def mse(a, b):
    return np.sum(np.power(a-b, 2))/len(a)
```

```python
def rmse(a, b):
    return np.sqrt(np.sum(np.power(a-b, 2)))
```

```python
num_of_test_bases = 20

cache_path = 'cache/3-completed-bases.npy'
if os.path.exists(cache_path) and not clear_cache:
    completed_bases = np.load('cache/3-completed-bases.npy')
    test_geometric_coefficients = np.load('cache/3-geometric-coefficients.npy')
    test_appearance_coefficients = np.load('cache/3-appearance-coefficients.npy')
    test_reconstructed_image = np.load('cache/3-reconstructed-image.npy')
    test_ncc = np.load('cache/3-ncc.npy')
    test_mse = np.load('cache/3-mse.npy')
    test_rmse = np.load('cache/3-rmse.npy')
else:
    completed_bases = np.zeros((num_of_test_bases,))
    test_geometric_coefficients = np.zeros((num_of_test_bases, num_of_test_samples, num_of_test_bases, 2))
    test_appearance_coefficients = np.zeros((num_of_test_bases, num_of_test_samples, num_of_test_bases))
    test_reconstructed_image = np.zeros((num_of_test_bases, num_of_test_samples, image_length))
    test_ncc = np.zeros((num_of_test_bases, num_of_test_samples,))
    test_mse = np.zeros((num_of_test_bases, num_of_test_samples,))
    test_rmse = np.zeros((num_of_test_bases, num_of_test_samples,))
    
bases = 1
while bases <= num_of_test_bases:
    print(bases)
    
    if completed_bases[bases-1] == 1:
        print('skipping {:}'.format(bases))
        bases += 1
        continue
    
    for i in range(num_of_test_samples):
        
        original_image = test_images[i]
        original_landmark = test_landmarks[i]
        
        warped_image = warp_image(original_image, image_shape,
                                  original_landmark, mean_landmark,
                                  anchor_borders=True, border_segments=8,
                                  interpolation_method='cubic')
        
        # compress geometry
        normalized_landmark = (original_landmark - mean_landmark) / std_landmark
        geometric_coefficients = [None] * 2
        for j in range(2):
            geometric_coefficients[j] = geometric_eig_vecs[j, :bases].dot(normalized_landmark[:, j])
        
        # compress appearance
        normalized_warped_image = (warped_image - mean_warped_image) / std_warped_image
        appearance_coefficients = appearance_eig_vecs[:bases].dot(normalized_warped_image)
        
        # uncompress appearance
        reconstructed_warped_image = appearance_eig_vecs[:bases].T.dot(appearance_coefficients) * std_warped_image + mean_warped_image
        
        # uncompress geometry
        reconstructed_landmark = np.zeros((num_of_landmarks, 2))
        for j in range(2):
            reconstructed_landmark[:, j] = geometric_eig_vecs[j, :bases].T.dot(geometric_coefficients[j])
        reconstructed_landmark = reconstructed_landmark * std_landmark + mean_landmark
        
        # reconstruct image
        reconstructed_image = warp_image(reconstructed_warped_image, image_shape,
                                         mean_landmark, reconstructed_landmark,
                                         anchor_borders=True, border_segments=8,
                                         interpolation_method='cubic')
        
        # save results
        test_geometric_coefficients[bases-1, i, :bases, 0] = geometric_coefficients[0]
        test_geometric_coefficients[bases-1, i, :bases, 1] = geometric_coefficients[1]
        test_appearance_coefficients[bases-1, i, :bases] = appearance_coefficients
        test_reconstructed_image[bases-1, i] = reconstructed_image
        test_ncc[bases-1, i] = ncc(original_image, reconstructed_image)
        test_mse[bases-1, i] = mse(original_image, reconstructed_image)
        test_rmse[bases-1, i] = rmse(original_landmark, reconstructed_landmark)
    
    completed_bases[bases-1] = 1
    bases += 1
            
    np.save('cache/3-completed-bases.npy', completed_bases)
    np.save('cache/3-geometric-coefficients.npy', test_geometric_coefficients)
    np.save('cache/3-appearance-coefficients.npy', test_appearance_coefficients)
    np.save('cache/3-reconstructed-image.npy', test_reconstructed_image)
    np.save('cache/3-ncc.npy', test_ncc)
    np.save('cache/3-mse.npy', test_mse)
    np.save('cache/3-rmse.npy', test_rmse)
```

## LDA

```python
def truncated_LDA(x, y):
    x_n = x[y == -1]
    x_p = x[y == +1]
    mean_n = np.mean(x_n, axis=0)
    mean_p = np.mean(x_p, axis=0)
    C = np.vstack((x_n - mean_n, x_p - mean_p)).T
    d, n = C.shape
    B = C.T.dot(C)
    U, L, V = np.linalg.svd(B)
    L = np.diag(L)
    A = np.zeros((d, n))
    for i in range(n):
        Ce = C.dot(V[:, i])
        A[:, i] = ( np.sqrt(L[i, i]) / np.linalg.norm(Ce) ) * Ce
    Y = A.T.dot(mean_n - mean_p)
    Z = np.linalg.solve((L * L).dot(V.T), Y)
    W = C.dot(Z)
    W = W / np.linalg.norm(W)
    return W
```
