## Principal Component Analysis

At this point we would create a covariance matrix:

$$
{\bf \Sigma} = {\bf X}^T {\bf X}
$$

However, that eats up far too much memory; in fact I can't compute it on my laptop. However, we only need the covariance matrix to obtain the first couple of eigenvectors as our principal axes of variation. We can use a trick be calculating the eigenvectors of a "psuedo" covariance matrix and transforming them to the eigenvectors we want.

$$
{\bf \Sigma} = {\bf X}^T {\bf X}, \quad {\bf \Sigma} {\bf e}_i = \lambda_i {\bf e}_i, \quad {\bf A} = {\bf X} {\bf X}^T \\
{\bf A} {\bf \alpha}_i = \lambda_i {\bf \alpha}_i \\
{\bf X} {\bf X}^T {\bf \alpha}_i = \lambda_i {\bf \alpha}_i \\
{\bf X}^T {\bf X} {\bf X}^T {\bf \alpha}_i = {\bf X}^T \lambda_i {\bf \alpha}_i \\
{\bf \Sigma} ( {\bf X}^T {\bf \alpha}_i ) = \lambda_i ( {\bf X}^T {\bf \alpha}_i) \\
{\bf \Sigma} {\bf e}_i = \lambda_i {\bf e}_i \\
{\bf e}_i = {\bf X}^T {\bf \alpha}_i
$$

```python
psuedo_covariance = norm_images.dot(norm_images.T)
```

Find psuedo eigenvectors and transform them into the real eigenvectors of the covariance matrix.

```python
junk, eig_vals, psuedo_eig_vecs = np.linalg.svd(psuedo_covariance)
eig_vecs = norm_images.T.dot(psuedo_eig_vecs.T).T
```

Normalize the eigenvectors because I want to. Their length isn't important since we are just going to project onto them anyway. Also, if they're normalized then they form an orthonormal bases which can be treated as a rotation.

```python
eig_vecs = np.apply_along_axis(lambda x: x/np.linalg.norm(x), 1, eig_vecs)
```

```python
num_of_bases = 20
trunc_eig_vecs = eig_vecs[:num_of_bases]
```

Things look funny. Let's try to see how the normalized cross-correlation behaves as we increase the number of eigenvectors used as a bases.

```python
def ncc(a, b):
    return np.sum( (a - np.mean(a)).dot(b - np.mean(b)) / (np.std(a) * np.std(b)) ) / len(a)
```

```python
def mse(a, b):
    return np.sum(np.power(a-b, 2))/len(a)
```

```python
num_of_test_bases = 40

test_ncc = np.zeros((num_of_test_bases, num_of_test_samples))
test_mse = np.zeros((num_of_test_bases, num_of_test_samples))

for bases in range(1, num_of_test_bases + 1):
    for i in range(num_of_test_samples):
        norm_test_image = (test_images[i] - mean_image) / std_image
        coeffs = eig_vecs[:bases].dot(norm_test_image)
        reconstructed_test_image = eig_vecs[:bases].T.dot(coeffs) * std_image + mean_image
        test_ncc[bases - 1, i] = ncc(test_images[i], reconstructed_test_image)
        test_mse[bases - 1, i] = mse(test_images[i], reconstructed_test_image)
```

```python
plt.figure(figsize=(3.125*2, 3.125*2), dpi=150)

plt.subplot(2, 1, 1)
for i in range(num_of_test_samples):
    plt.plot(np.arange(num_of_test_bases), test_ncc[:, i], alpha=0.25)
plt.plot(np.arange(num_of_test_bases), np.mean(test_ncc, axis=1), 'r-', linewidth=2)
plt.xlim(0, num_of_test_bases-1)
plt.plot([20, 20], plt.ylim(), 'k--')
plt.ylabel('NCC')
plt.title('Normalized Cross-Correlation')
    
plt.subplot(2, 1, 2)
for i in range(num_of_test_samples):
    plt.plot(np.arange(num_of_test_bases), test_mse[:, i], alpha=0.25)
plt.plot(np.arange(num_of_test_bases), np.mean(test_mse, axis=1), 'r-', linewidth=2)
plt.xlim(0, num_of_test_bases-1)
plt.plot([20, 20], plt.ylim(), 'k--')
plt.xlabel('Number of Eigenvectors ($k_I$)')
plt.ylabel('MSE')
plt.title('Mean Squared Error')

plt.savefig('report/1-error.png', dpi=150, bbox_inches='tight')
```
