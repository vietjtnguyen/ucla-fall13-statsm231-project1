```python
def generate_grid_points(xs, ys):
    m = np.zeros((len(xs)*len(ys), 2))
    for i in range(len(ys)):
        m[len(xs)*i:len(xs)*(i+1), 0] = xs
        m[len(xs)*i:len(xs)*(i+1), 1] = np.ones((len(xs),)) * ys[i]
    return m
```

```python
def warp_image(image, image_shape, from_landmarks, to_landmarks, anchor_borders=True, border_segments=2, interpolation_method='linear'):
    
    if anchor_borders:
        segments = border_segments
        borders = np.zeros((segments * 4, 2))
        borders[segments*0:segments*1, 0] = np.linspace(0, 1, segments, False) * image_shape[0]
        borders[segments*0:segments*1, 1] = 0
        borders[segments*1:segments*2, 0] = image_shape[0]
        borders[segments*1:segments*2, 1] = np.linspace(0, 1, segments, False) * image_shape[1]
        borders[segments*2:segments*3, 0] = (1 - np.linspace(0, 1, segments, False) ) * image_shape[0]
        borders[segments*2:segments*3, 1] = image_shape[1]
        borders[segments*3:segments*4, 0] = 0
        borders[segments*3:segments*4, 1] = (1 - np.linspace(0, 1, segments, False) ) * image_shape[1]
        border_deltas = np.zeros((segments * 4, 2))
        
    grid_points = generate_grid_points(np.arange(image_shape[0]), np.arange(image_shape[1]))
        
    deltas = to_landmarks - from_landmarks
    if anchor_borders:
        interp_points = np.vstack((from_landmarks, borders))
        deltas = np.concatenate((deltas, border_deltas))
    else:
        interp_points = from_landmarks
        deltas = deltas
        
    interp_delta = sp.interpolate.griddata(interp_points,
                                           deltas,
                                           grid_points,
                                           method=interpolation_method,
                                           fill_value=0)
    
    orig_pixels = grid_points
    disp_pixels = orig_pixels + interp_delta
    warped_image = sp.interpolate.griddata(disp_pixels,
                                           image,
                                           orig_pixels,
                                           method=interpolation_method,
                                           fill_value=0)
    
    return warped_image
```
